# Parse the lot

```python
full_result = None
failed = []
for n,row in working.iterrows():
    print "-------"
    print n
    print "URL: {}".format(row.Link)
    if not pd.isnull(row.Link) and row.Link != "Not available":
        try:
            table_df = clean.parse_row(row)
            if table_df is not None:
                print "\t{} rows added".format(table_df.shape[0])
                if full_result is None:
                    full_result = table_df
                else:
                    full_result = pd.concat([full_result, table_df])
            else:
                print "!!! - 0 rows added"
        except Exception as e:
            print "Failed with {}".format(e) 
            failed.append((n, row.Link, e))
```
