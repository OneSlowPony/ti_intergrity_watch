```python
BASE_URL = "https://www.gov.uk"

def url_for_department_search(department, search_term):
    return (
        BASE_URL + 
        "/government/publications.json?"
        "keywords={}&"
        "publication_filter_option=transparency-data&"
        "topics[]=all&"
        "departments[]={}&"
        "official_document_status=all&"
        "world_locations[]=all&"
    ).format(search_term, department)

def json_response(url):
    proxies = {
      'http': 'thproxy.internet.point:8080',
      'https': 'thproxy.internet.point:8080',
    }
    
    r = requests.get(url, proxies=proxies)
    
    return r.json() 
```
